```python
#데이터 로드 청크 분할
from langchain_community.document_loaders import Docx2txtLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = Docx2txtLoader("/Users/kshi3430/무제 폴더/제8회 교육 공공데이터 AI 활용대회 운영 안내.docx")  # 문서 로더 초기화
docs = loader.load() 

text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=1000,
    chunk_overlap=200,
)

documents = text_splitter.split_documents(docs)
print(len(documents))
for i in range(10):
    print(documents[i].page_content)
```

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# pinecone 연결
from langchain_upstage import UpstageEmbeddings
from langchain_pinecone import PineconeVectorStore
from pinecone import Pinecone


 
embeddings = UpstageEmbeddings(model="solar-embedding-1-large")
index_name="pratice-langchain"
database = PineconeVectorStore.from_documents(documents, embeddings,index_name=index_name)
```

```python
query = "공공대회 ai 활용대회는 기획만 하는건가요? 아니면 구현까지 해야하나요?"

retriever = database.as_retriever(search_kwargs={"k": 3})

docs = retriever.invoke(query)
print(len(docs))
for i in range(len(docs)):
    print(docs[i].page_content)
```
